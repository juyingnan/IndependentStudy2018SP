```python
import matplotlib.pyplot as plt
from skimage.color import rgb2hed
from matplotlib.colors import LinearSegmentedColormap
from skimage import io
```

```python
image_path1 = 'c:/Users/bunny/Desktop/test2.jpg'
ihc_rgb = io.imread(image_path1)
ihc_hed = rgb2hed(ihc_rgb)
```

```python
import numpy as np
from skimage.exposure import rescale_intensity
```

```python
# Rescale hematoxylin and DAB signals and give them a fluorescence look
h = rescale_intensity(ihc_hed[:, :, 0], out_range=(0, 1.0))
e = rescale_intensity(ihc_hed[:, :, 1], out_range=(0, 1.0))
d = rescale_intensity(ihc_hed[:, :, 2], out_range=(0, 1.0))
zdh = np.dstack((e, h, (np.zeros_like(h))))
```

```python
%matplotlib inline
fig = plt.figure()
plt.imshow(zdh)
plt.show()
```

```python
zdh_2 = np.dstack((h, e, d))
```

```python
%matplotlib inline
fig = plt.figure()
plt.imshow(zdh_2)
plt.show()
```
